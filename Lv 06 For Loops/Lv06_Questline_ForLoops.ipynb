Die erste gerade Zahl wre also die 2 und die letzte die 100.\
d.h. 2 + 4 + 6 + 8 +10 ... + 98 + 100\
Wichtig: In der Konsolenausgabe sollte nur eine print-Anweisung stehen.\

Hinweis: Es gibt mehrere Mglichkeiten, dieses Problem zu lsen,\

<div class="alert alert-block alert-success">
<b>Lsungen Lv 6 Adding Numbers</b>
</div>

```python
# Adding Even Numbers
even_result = 0

for number in range(1, 101):
    if number % 2 == 0:
        even_result += number
print(even_result)
```

```python
# Adding Odd Numbers
odd_result = 0

for number in range(1, 101):
    if number % 2 != 0:
        odd_result += number

print(odd_result)
```

#### **Lv 5 Quest 4 FizzBuzz**

Schreiben Sie ein Programm, dass die Zahlen 1 bis 100 in der Konsole ausgibt\
Aber\
fr jede Zahl die durch 3 teilbar ist, soll Fizz ausgegeben werden\
fr jede Zahl die durch 5 teilbar ist, soll Buzz ausgegeben werden\
und fr jede Zahl die durch 3 und durch 5 teilbar ist, soll FizzBuzz ausgegeben werden.


<div class="alert alert-block alert-success">
<b>Lsungen Lv 6 Fizz Buzz</b>
</div>

```python
for number in range(1, 101):
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)
```
