## **Vergleichsoperatoren | Binary Operator**
binary Operators using left-sided binding\
Vergleichsoperatoren werden eingesetzt um Werte zu vergleichen.\
Auf der Grundlage des Vergleichs werden dann bestimmte Entscheidungen getroffen.

| Operator | Function | Funktion
|:---------:|:-------------:|:-------------:|
| <span style="color:#f5dd">===</span> | is equal to | ist gleich |
| <span style="color:#f5dd">!==</span> | is not equal to | ist ungleich |
| <span style="color:#f5dd"> > </span> | is greater than | größer als |
| <span style="color:#f5dd"> < </span> | is lesser than | kleiner als |
| <span style="color:#f5dd"> >= </span> | is greater or equal to | größer gleich |
| <span style="color:#f5dd"> <= </span> | is lesser or equal to | kleiner gleich |

```python
# <                 Kleiner als
# >                 Größer als
# <=                Kleiner gleich
# >=                Größer gleich
# ==                ist gleich
# !=                ist ungleich
# <<, >>            Bitweise Verschiebung
# ^=, |=, >>=, <<=

# Beispiele
7 == 7  # True
7 == 7. # True
7 == 14 # False

var = 0
print(var == 0)
print(var != 0)

var = 1
print(var == 0)
print(var != 0)
```

### **Beispiele**

**Muss ich mich warm anziehen?**

```python
grad_celsius_outside = 3

print(grad_celsius_outside <= 10.0)  # Lower than or equal to
```

**Wird der Blitzer ausgelöst?**\
**Ergebnisse aus einem Vergleich können in Variablen gespeichert werden**

```python
current_velocity_kmh = 90

# currently_velocity_kmh < 85
result = current_velocity_kmh >= 85
print(result)
print("Ja der Blitzer wird ausgelöst.")
```

## **Logische Operatoren**
Logische Operatoren werden verwendet um den Programmablauf mittels Logik zu steuern.

```python
# && # and # Konjunktion
# || # or # Disjunktion
# not # Negation

var = 1
# Beispiel 1:
print(var > 0)
print(not (var <= 0))


# Beispiel 2:
print(var != 0)
print(not (var == 0))
```
