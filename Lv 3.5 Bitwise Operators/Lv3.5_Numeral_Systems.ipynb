# **Lv 3.5 Number Systems**

## **Umrechnung dezimaler Zahlen in Binärzahlen**

```python
# hier einmal positiv und negativ darstellen
# und erläutern warum diese Darstellung nicht korrekt ist
zahl = -7
binär = bin(zahl)
print(f"Die binäre Darstellung von {zahl} ist standardmäßig in Python: {binär}")
```

## **Einer Komplementär**

[Darstellung negativer Binärzahlen Youtube](https://www.youtube.com/watch?v=4qH4unVtJkE)

```python
# funktioniert ebenfalls nicht 
zahl = ~5
binär = bin(zahl)
print(f"Die binäre Darstellung von {zahl} ist: {binär}")
```

## **Zweier Komplementär**

[Binary Addition and Subtraction & Complements](https://www.youtube.com/watch?v=sJXTo3EZoxM&t=804s)

```python
# Zweierkomplementär --> korrekte Darstellung
number = -7
byte_number = number.to_bytes(1, byteorder="little", signed=True)
binary_number = bin(int.from_bytes(byte_number, byteorder="little"))
print(f"Die binäre Darstellung von {number} ist:, {binary_number}")
```
