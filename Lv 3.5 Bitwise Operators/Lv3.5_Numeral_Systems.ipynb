# **Lv 3.5 Number Systems**

## **Umrechnung dezimaler Zahlen in Binrzahlen**

```python
# hier einmal positiv und negativ darstellen
# und erlutern warum diese Darstellung nicht korrekt ist
zahl = -7
binr = bin(zahl)
print(f"Die binre Darstellung von {zahl} ist standardmig in Python: {binr}")
```

## **Einer Komplementr**

[Darstellung negativer Binrzahlen Youtube](https://www.youtube.com/watch?v=4qH4unVtJkE)

```python
# funktioniert ebenfalls nicht 
zahl = ~5
binr = bin(zahl)
print(f"Die binre Darstellung von {zahl} ist: {binr}")
```

## **Zweier Komplementr**

[Binary Addition and Subtraction & Complements](https://www.youtube.com/watch?v=sJXTo3EZoxM&t=804s)

```python
# Zweierkomplementr --> korrekte Darstellung
number = -7
byte_number = number.to_bytes(1, byteorder="little", signed=True)
binary_number = bin(int.from_bytes(byte_number, byteorder="little"))
print(f"Die binre Darstellung von {number} ist:, {binary_number}")
```
