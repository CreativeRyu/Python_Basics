### **Return zum Verlassen einer Funktion**

```python
def fill_hearts(hearts = 0):
    if hearts == 0:
        return
    else:
        print(f"Got filled up your hearts by {hearts}")

fill_hearts()
fill_hearts(2)
```

### **Return als Rückgabe des Ergebnisses einer Funktion**

```python
def format_name(f_name, l_name):
    f_name = f_name.title()
    l_name = l_name.title()
    # print(f"{f_name} {l_name}")
    return f"{f_name} {l_name}"

print(format_name("ryu", "rONin"))
```

### **Rückgabe mehrere Werte in ein Tuple**

```python
def spam():
    wert = 42
    return wert * 2, wert / 2, "Auch ich werde zurück gegeben."

great_value, small_value, some_text = spam()
tup1 = spam()
print(great_value)
print(small_value)
print(some_text)
print(tup1)

# print(type(great_value))
# print(type(small_value))
# print(type(some_text))

# Das gesamte Tupel wird in einer Variable vom Typ Tupel gespeichert
one_variable = spam()
print(type(one_variable))
```

### **Return am Beispiel Odd or Even**

```python
def odd_or_even(rest):
    match rest:
        case 0:
            return "Es handelt sich um eine gerade Zahl."
        case _:
            return "Es handelt sich um eine ungerade Zahl."

number = 5
rest = number % 2

result = odd_or_even(rest)
print(result)
```

```python
def is_leap(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False

def days_in_month(year,month):
    if month == 2 and is_leap(year):
        return 29
    else:
        month_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        return month_days[month-1]


#🚨 Do NOT change any of the code below 
year = int(input("Enter a year: "))
month = int(input("Enter a month: "))
days = days_in_month(year, month)
print(days)
```
