# **Lambda | Anonymous Function**

Warum Lambda?
Es gibt Funktionen die eine Funktion als Argument übergeben bekommen wollen.

```python
# Lambdas sind einfache kurze Funktionen, diese können einer Variable übergeben werden.

a = lambda x: x + 1
print(a(10))
```

```python
simple_calc = lambda a,b: a + b
print(simple_calc(2,3))
```

```python
# exercise
# create a lambda function that accepts one integer argument
# if the integer is > 5 return hello
# otherwise return bye

tachi = lambda x: "hello" if x > 5 else "bye"
print(tachi(7))
```

```python
lst_comp = [x ** 2 for x in range(10)]
print(lst_comp)

lst = [1, 2, 3, 4, 5, 6]

# Funktion mit der gleichen Aufgabe
def square_value(val):
    return val**2

# Mapping mit der entsprechenden Funktion
lst_squared = list(map(square_value, lst))

print(lst_squared)
list_lambda_squared = list(map(lambda x: x**2, lst))
print(list_lambda_squared)
```

```python
# Aufbau eines Lambdas

any_value = 10
# print((lambda x: x * 4 + 1)(any_value))
unser_lambda = (lambda x: x * 4 + 2)
print(unser_lambda(any_value))
```

```python
# einer Variable eine Lambda Expression zu weisen

spam = lambda x, y: x * y + 2
print(spam(4,10))
```

```python
import functools
irgendeine_liste = [10, -2, 5, -10, 7, 23, -1]
list_max = functools.reduce(lambda x, y: x if x > y else y, irgendeine_liste)
print(list_max)
list_min = functools.reduce(lambda x, y: x if x < y else y, irgendeine_liste)
print(list_min)
```

```python
filtered_list = list(filter(lambda x: True if x >= 0 else False, irgendeine_liste))
print(filtered_list)
```
