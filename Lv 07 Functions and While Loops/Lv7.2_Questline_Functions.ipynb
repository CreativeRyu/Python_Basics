#### **Lv 7.2 Quest 1 Fixed Hurdles**


1. Funktionen erläutern
2. Ziel erreichen durch Funktionen erreichen lassen
3. Einen For Loop verwenden
4. Einen While Loop statt des For Loops einsetzen

<div class="alert alert-block alert-success">
<b>Lösung Lv 7.2 Quest 1 Fixed Hurdles</b>
</div>

```python
# Lösung For Loop
for step in range(6):
    jump()

# Lösung While Loop
number_of_hurdles = 6
while number_of_hurdles > 0:
    jump()
    number_of_hurdles -= 1
    print(number_of_hurdles)
```
