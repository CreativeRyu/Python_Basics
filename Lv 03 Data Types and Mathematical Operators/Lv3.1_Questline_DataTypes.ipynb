## **Konvertierung von Datentypen**

```python
# Datentyp Konvertierung in einen String
num_string1 = input("Gib eine Zahl ein:\n")
num_string2 = input("Gib noch eine Zahl ein. \n")
result = num_string1 + num_string2 

print("Das Ergebnis der beiden Zahlen ist {result}")
print(type(result))

num1 = int(num_string1)
num2 = int(num_string2)

numeric_result = num1 + num2
print("Das Ergebnis der beiden Zahlen ist {numeric_result}")
print(type(numeric_result))
```
