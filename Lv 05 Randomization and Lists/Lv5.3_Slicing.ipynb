# **Lv 5.3 Slicing**

```python
piano_keys = ["a","b","c","d","e","f","g"]
print(piano_keys[2:5])
```

```python
# Gibt die Liste ab der Position aus
print(piano_keys[2:])

# Gibt die Liste bis zur Position aus
print(piano_keys[:5])

# Gibt jeden zweiten Wert im Slice zurck (Inkrement)
print(piano_keys[2:5:2])

# Gibt jeden zweiten Wert zurck
print(piano_keys[::2])

# Invertiert die Liste | Reverses the List
print(piano_keys[::-1])

# List Comprehensions bzw. Listen knnen auch bei Minuswerten beginnen
print([i for i in range(-1, 4)])
```
